```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
# import data
data0=pd.read_csv('iris.csv').values
data1=pd.read_csv('dataI.csv').values
data2=pd.read_csv('dataII.csv').values
data3=pd.read_csv('dataII.csv').values
data4=pd.read_csv('dataIV.csv').values
data5=pd.read_csv('dataV.csv').values
```

```python
# general procedure
SS = StandardScaler().fit(data0)    # get mean & variance using data0
X = SS.transform(data0)             # standardize data using mean & variance of data0
pca = PCA(n_components=2).fit(X)    # fit PCA to standardized data
Tr = pca.transform(X)               # transform data to PCs using cov of X
reconstr_data=pca.inverse_transform(Tr)
```
